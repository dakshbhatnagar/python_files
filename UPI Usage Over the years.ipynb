```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('UPIData.csv')
```

```python
df
```

```python
df1 = df.set_index('Month-Year',inplace=True)
```

```python
VolTxn = df['VolumeOfTransaction_millions'].sum()
VolTxn
```

```python
print('The Sum of the volume of transactions between 2016 and 2020 is {} Millions'.format(int(VolTxn)))
```

```python
plt.figure(figsize=(14, 7))
plt.plot(df.Year,df.VolumeOfTransaction_millions)
plt.plot(df.Year,df.AmountOfTransactionInBn)
plt.xlabel('AmountOfTransactionInBn')
plt.ylabel('Volume of Txn M')
plt.title('UPI usage over the years')
plt.legend(['Volume Of Transactions','Amount Of transactions']);
```

```python
plt.figure(figsize=(14, 7))
sns.scatterplot(x=df.BanksUsingUPI, y=df.VolumeOfTransaction_millions, hue = df.Year);
```

```python
plt.figure(figsize=(14, 7))
sns.scatterplot(x=df.BanksUsingUPI, y=df.AmountOfTransactionInBn, hue = df.Year);
```

```python
sns.pairplot(df, hue='Year');
```

From the analysis above, what you can see is the banks using UPI has decreased over the years however the volume of transactions and the amount of transactions has gone uphill which means there has been a great amount of acceptance by the common public of the digital payments technology.
