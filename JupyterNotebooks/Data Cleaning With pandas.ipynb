```python
FILE_PATH = 'https://github.com/AlexTheAnalyst/PandasYouTubeSeries/raw/main/Customer%20Call%20List.xlsx'
```

```python
import pandas as pd
```

```python
#!pip install openpyxl --quiet
```

```python
df = pd.read_excel(FILE_PATH)
```

```python
df.head()
```

```python
df[["Street_Address", "State", "Zip_Code"]]  = df["Address"].str.split(',',expand=True)
```

```python
df.loc[df['Paying Customer']=='N','Paying Customer'] = 'No'
df.loc[df['Paying Customer']=='Y','Paying Customer'] = 'Yes'
df.loc[df['Paying Customer']=='N/a','Paying Customer'] = None
```

```python
df.loc[df['Do_Not_Contact']=='Y','Do_Not_Contact'] = 'Yes'
df.loc[df['Do_Not_Contact']=='N','Do_Not_Contact'] = 'No'
```

```python
df['Phone_Number'] = [str(i).replace("/","-") for i in df['Phone_Number']]
df['Phone_Number'] = [str(i).replace("|","-") for i in df['Phone_Number']]
```

```python
mask = ~df['Phone_Number'].str.contains("-")
df.loc[mask, 'Phone_Number']  = df.loc[mask, 'Phone_Number'].apply(lambda x : x[0:3] + "-" + x[3:6] + "-" + x[6:])
```

```python
mask = df['Phone_Number'].str.contains("n|N")
df.loc[mask, 'Phone_Number'] = None
```

```python
df = df[['CustomerID', 'First_Name', 'Last_Name', 'Phone_Number', 'Street_Address', 'State', 'Zip_Code',
       'Paying Customer', 'Do_Not_Contact']]
```

```python
df['Last_Name'] = df['Last_Name'].str.replace('/|\.|_', "", regex=True)
```
