2. Measures of dispersion

Measures of dispersion also known as measures of variability or spread, are statistical metrics that describe the extent to which data points in a dataset vary or spread out from the central tendency (such as the mean, median, or mode). These measures provide valuable insights into the distribution and variability of data, helping us understand how data points are dispersed or clustered.

Common measures of dispersion include:

a. **Range:** The range is the simplest measure of dispersion and is calculated as the difference between the maximum and minimum values in the dataset. It provides an idea of the spread of data.

   $$Range = \text{Maximum Value} - \text{Minimum Value}$$

````python
# Dataset
a = [104, 98, 90, 104, 104]

# Find the maximum and minimum values in the dataset
max_value = max(a)
min_value = min(a)

# Calculate the range
range_value = max_value - min_value

print(f"Maximum value in the data: {max_value}")
print(f"Minimum value in the data: {min_value}")
print(f"Range of the data: {range_value}")
````

b. **Variance:** Variance quantifies the average of the squared differences between each data point and the mean. It provides information about how individual data points deviate from the mean.

   Population Variance 
   
   $$\sigma^2 = \frac{\sum_{i=1}^{N}(x_i - \mu)^2}{N}$$
   
   Sample Variance 
   
   $$s^2 = \frac{\sum_{i=1}^{N}(x_i - \bar{x})^2}{N - 1}$$


   Where:
   - x is an individual data point.
   - μ is the population mean.
   - x̄ is the sample mean.
   - N is the total number of data points.

````python
#How is variance Calculated
# Dataset
a = [104, 98, 90, 104, 104]

# Calculate the mean
a_mean = np.mean(a)
print(f'Mean: {a_mean:.2f}')

# Calculate the differences between each item and the mean
differences = [i - a_mean for i in a]
print(f'Differences from the mean: {differences}')

# Calculate the squared differences
squared_differences = [(i - a_mean) ** 2 for i in a]
print(f'Squared differences: {squared_differences}')

# Calculate the variance (average of squared differences)
variance = np.mean(squared_differences)
print(f'Variance: {variance:.2f}')
````

c. **Standard Deviation:** 

The standard deviation is the square root of the variance. It measures the average amount of variation or dispersion in the data. A smaller standard deviation indicates less variability. Population Standard Deviation and Sample Standard Deviation are two statistical measures used to quantify the spread or dispersion of data points in a dataset. They are similar in concept but are calculated differently and serve distinct purposes:

**Population Standard Deviation (σ):**

1. **Calculation:** The population standard deviation (σ) is calculated by taking the square root of the average of the squared differences between each data point in the entire population and the population mean (μ). The formula is:

   $$\sigma = \sqrt{\frac{Σ(x_i - μ)^2}{N}}$$

   where:
   - $σ$ is the population standard deviation.
   - $x_i$ represents each individual data point in the population.
   - $μ$ is the population mean.
   - $N$ is the total number of data points in the population.

2. **Use:** The population standard deviation is used when you have data for an entire population, meaning you have information about every member of the group you are studying. It provides a measure of how data points in the entire population vary from the population mean.

**Sample Standard Deviation (s):**

1. **Calculation:** The sample standard deviation (s) is calculated similarly to the population standard deviation but with one key difference. Instead of dividing by the total number of data points (N), it divides by (N - 1). The formula is:

   $$s = \sqrt{\frac{Σ(x_i - \bar{x})^2}{N-1}}$$

   where:
   - $s$ is the sample standard deviation.
   - $x_i$ represents each individual data point in the sample.
   - $\bar{x}$ is the sample mean.
   - $N$ is the total number of data points in the sample.

2. **Use:** The sample standard deviation is used when you have a subset or sample of data from a larger population. It provides an estimate of how data points in the sample vary from the sample mean and is used to make inferences about the population from which the sample was drawn. The use of (N - 1) in the denominator (known as Bessel's correction) corrects for bias in the estimation of population standard deviation when working with a sample.

**Key Differences:**

1. **Formula:** The formulas for population and sample standard deviations are very similar, but the denominator differs: σ divides by N (population size), while s divides by (N - 1) (sample size minus 1).

2. **Purpose:** Population standard deviation is used when you have data for an entire population, whereas sample standard deviation is used when you have data from a sample and want to estimate the population standard deviation.

3. **Bias Correction:** Sample standard deviation includes Bessel's correction (dividing by N - 1) to account for the fact that you are estimating population parameters from a sample, which introduces bias if not corrected.

````python
#How is Standard Deviation Calculated
# Dataset
a = [104, 98, 90, 104, 104]

# Calculate the mean
a_mean = np.mean(a)
print(f'Mean: {a_mean:.2f}')

# Calculate the differences between each item and the mean
differences = [i - a_mean for i in a]
print(f'Differences from the mean: {differences}')

# Calculate the squared differences
squared_differences = [(i - a_mean) ** 2 for i in a]
print(f'Squared differences: {squared_differences}')

# Calculate the variance (average of squared differences)
variance = np.mean(squared_differences)
print(f'Variance: {variance:.2f}')

# Calculate the standard deviation (square root of variance)
std_deviation = np.sqrt(variance)
print(f'Standard Deviation: {std_deviation:.2f}')
````

d. **Interquartile Range (IQR):** The Interquartile Range (IQR) is a statistical measure that uses quartiles, specifically the first quartile (Q1) and the third quartile (Q3), to describe the spread or variability of data. It is a robust measure of spread because it is not influenced by extreme values or outliers in the dataset.

To understand how the IQR is calculated using quartiles and the median, let's break down the process step by step:

1. **Data Arrangement:** Begin with a dataset of values, which you can arrange in ascending order from smallest to largest.

2. **Calculate the Median (Q2):** Find the median (Q2) of the dataset. The median is the middle value when the data is sorted. If there's an odd number of data points, the median is the middle value. If there's an even number of data points, the median is the average of the two middle values.

3. **Split the Data into Two Parts:** Divide the dataset into two parts:
   - The lower half, which includes all values less than or equal to the median (Q2).
   - The upper half, which includes all values greater than or equal to the median (Q2).

4. **Calculate the First Quartile (Q1):** Find the median (Q1) of the lower half of the dataset. This is the median of the values below the overall median (Q2).

5. **Calculate the Third Quartile (Q3):** Find the median (Q3) of the upper half of the dataset. This is the median of the values above the overall median (Q2).

6. **Calculate the IQR:** The Interquartile Range (IQR) is the range between the first quartile (Q1) and the third quartile (Q3). It represents the spread of the middle 50% of the data and is calculated as follows:
   
   $$IQR = Q3 - Q1$$

The IQR is a valuable measure of variability because it is not affected by outliers or extreme values. It focuses on the central portion of the data distribution, providing insights into how the middle values are dispersed. A larger IQR indicates greater spread or variability within the middle portion of the data, while a smaller IQR suggests that the data is more concentrated around the median.

````python
# Dataset without outliers
data_without_outliers = [104, 98, 90, 104, 104]

# Calculate the first quartile (Q1) for the dataset without outliers
Q1_without_outliers = np.percentile(data_without_outliers, 25)

# Calculate the third quartile (Q3) for the dataset without outliers
Q3_without_outliers = np.percentile(data_without_outliers, 75)

# Calculate the Interquartile Range (IQR) for the dataset without outliers
IQR_without_outliers = Q3_without_outliers - Q1_without_outliers

# Print Q1, Q3, and IQR for the dataset without outliers
print("Dataset without outliers:")
print("Q1:", Q1_without_outliers)
print("Q3:", Q3_without_outliers)
print("Interquartile Range (IQR):", IQR_without_outliers)

# Dataset with outliers
data_with_outliers = [104, 98, 90, 104, 1040]

# Calculate the first quartile (Q1) for the dataset with outliers
Q1_with_outliers = np.percentile(data_with_outliers, 25)

# Calculate the third quartile (Q3) for the dataset with outliers
Q3_with_outliers = np.percentile(data_with_outliers, 75)

# Calculate the Interquartile Range (IQR) for the dataset with outliers
IQR_with_outliers = Q3_with_outliers - Q1_with_outliers

# Print Q1, Q3, and IQR for the dataset with outliers
print("\nDataset with outliers:")
print("Q1:", Q1_with_outliers)
print("Q3:", Q3_with_outliers)
print("Interquartile Range (IQR):", IQR_with_outliers)
````

## Probability

Probability is a fundamental concept in mathematics and statistics that quantifies uncertainty or the likelihood of an event occurring. It provides a way to describe and analyze uncertain situations, make predictions, and make informed decisions in the presence of randomness.

**Definition of Probability:**

The probability of an event is a numerical measure that ranges from 0 to 1, where:
- 0 indicates that the event is impossible (it will not occur).
- 1 indicates that the event is certain (it will occur with certainty).
- Values between 0 and 1 represent the degree of uncertainty or likelihood of the event occurring.

In mathematical terms, the probability of an event E, denoted as P(E), is defined as:

$$P(E) = \frac{\text{Number of Favorable Outcomes}}{\text{Total Number of Possible Outcomes}}$$

**Classic Probability Model (Theoretical Probability):**

The classic probability model, also known as theoretical or a priori probability, is based on the fundamental principles of probability theory. It assumes that we have complete knowledge of the experiment and can calculate probabilities from first principles. In this model:

- The sample space (S) is the set of all possible outcomes of an experiment.
- Each outcome in the sample space is assumed to be equally likely (equiprobable) when there is no reason to believe otherwise.
- Probabilities are determined by counting the number of favorable outcomes for an event and dividing by the total number of possible outcomes.

For example, when rolling a fair six-sided die, each face (1, 2, 3, 4, 5, and 6) is equally likely to occur. Therefore, the probability of rolling a specific number, say 3, is 1/6.

**Relative Frequency of Occurrence (Empirical Probability):**

The relative frequency of occurrence, also known as empirical probability, is based on observations and data collected from actual experiments or events. In this model:

- Probabilities are estimated by conducting repeated trials or observations of the experiment.
- The relative frequency of an event is the ratio of the number of times the event occurs to the total number of trials or observations.

*For example*, in a series of coin flips, we can estimate the probability of getting heads by counting how many times heads occurred in the trials and dividing by the total number of trials.

> The key distinction between the classic and relative frequency models is that the classic model relies on theoretical assumptions of equal likelihood, while the relative frequency model is based on observed data. In practice, empirical probability is often used when it's challenging to make theoretical assumptions about probabilities.

Both models are valuable tools for understanding and dealing with uncertainty in different contexts, and they are fundamental to the field of probability and statistics.

Let's explore the key concepts like events, trials, outcomes, and probability itself:

**1. Experiment:**
   - An experiment is a process or procedure that generates an outcome. It can be any action or situation where there is uncertainty about the result.

**2. Outcome:**
   - An outcome is a possible result or observation that can occur as a result of an experiment. Outcomes can be either simple (a single result) or compound (a combination of results).

   - Probability is a numerical measure of the likelihood of an event occurring. It quantifies the uncertainty associated with the outcomes of an experiment. Probability values range from 0 (impossible event) to 1 (certain event).

What you saw above is a **Venn Diagram** which is a graphical representation used to visually illustrate the relationships and intersections between different sets or groups and is widely used in Probability. It was developed by the British mathematician and philosopher John Venn in the late 19th century. Venn diagrams are a valuable tool in various fields, including mathematics, statistics, logic, and data analysis, for organizing and comparing data sets.

Here are key components and characteristics of Venn diagrams:

**1. Circles or Ellipses:** Venn diagrams typically consist of two or more circles or ellipses, each representing a set or group. These circles can overlap to show the relationships between the sets.

**2. Intersections:** The overlapping regions between circles represent the elements that are common to the sets they intersect. The size of the overlapping area corresponds to the size of the intersection.

**3. Unique Regions:** The non-overlapping regions within each circle represent the elements that are unique to that particular set. These regions are often labeled with the elements or characteristics that belong exclusively to that set.

**4. Universal Set:** In some cases, a rectangle or another shape is used to enclose all the circles, representing the universal set that contains all possible elements under consideration.

**5. Set Labels:** Sets are labeled with names or identifiers to indicate what they represent. These labels are often placed inside or beside the circles.

**6. Purpose:** Venn diagrams are used to visually depict relationships and differences between sets, highlighting commonalities and distinctions. They are especially useful for understanding set theory, logic, and categorical data.

**Uses:**
   - **Set Theory:** Venn diagrams are commonly used to illustrate set operations such as union, intersection, and complement.
   - **Statistics:** In data analysis, Venn diagrams can be used to show how different data sets overlap or share common elements.
   - **Logic:** Venn diagrams help illustrate logical relationships between statements and propositions.
   - **Probability:** They can be used to visualize events and their intersections in probability theory.
   - **Euler Diagrams:** Euler diagrams are a specific type of Venn diagram that can represent more complex relationships, allowing for multiple set intersections.

**Limitations:** 

While Venn diagrams are effective for representing relationships between a small number of sets, they become less practical as the number of sets increases. Overlapping areas can become cluttered and difficult to interpret in complex diagrams.

## Union and intersection

Union and intersection are fundamental set operations used in set theory and mathematics to describe relationships between sets. They have important applications in various fields, including probability, statistics, and computer science.

**Union (⋃):**

The union of two or more sets is an operation that combines all the unique elements from those sets into a single set. It is denoted by the symbol "⋃" and is defined as follows:

- Given sets A and B, the union of A and B (A ⋃ B) contains all elements that are in A, in B, OR in both.
- Mathematically, for any element x:
  - x ∈ (A ⋃ B) if and only if (x ∈ A) or (x ∈ B).

For example, if A = {1, 2, 3} and B = {3, 4, 5}, then A ⋃ B = {1, 2, 3, 4, 5}. It includes all unique elements from both sets.

**Intersection (⋂):**

The intersection of two or more sets is an operation that identifies the elements that are common to all those sets. It is denoted by the symbol "⋂" and is defined as follows:

- Given sets A and B, the intersection of A and B (A ⋂ B) contains all elements that are both in A and in B.
- Mathematically, for any element x:
  - x ∈ (A ⋂ B) if and only if (x ∈ A) and (x ∈ B).

For example, if A = {1, 2, 3} and B = {3, 4, 5}, then A ⋂ B = {3}. It includes only the element that is common to both sets.

**Properties and Use Cases:**

1. **Commutative:** Both union and intersection operations are commutative, meaning the order of the sets does not matter:
   - A ⋃ B = B ⋃ A
   - A ⋂ B = B ⋂ A

2. **Associative:** Both operations are also associative, meaning the grouping of sets does not affect the result:
   - (A ⋃ B) ⋃ C = A ⋃ (B ⋃ C)
   - (A ⋂ B) ⋂ C = A ⋂ (B ⋂ C)

3. **Distributive:** Union and intersection distribute over each other:
   - A ⋃ (B ⋂ C) = (A ⋃ B) ⋂ (A ⋃ C)
   - A ⋂ (B ⋃ C) = (A ⋂ B) ⋃ (A ⋂ C)

4. **De Morgan's Laws:** De Morgan's laws describe how union and intersection interact with complements:
   - (A ⋃ B)' = A' ⋂ B'
   - (A ⋂ B)' = A' ⋃ B'

5. **Use Cases:**
   - In probability, the union represents the occurrence of at least one of multiple events, and the intersection represents the occurrence of all events simultaneously.
   - In set theory, these operations are used to manipulate and analyze sets.
   - In database query languages like SQL, union and intersection operations are used to combine or filter data from multiple tables.

Understanding union and intersection is crucial in various mathematical and analytical contexts, as they help define relationships between sets and allow for complex set manipulations and analysis.

**6. Probability of an Event (P(E)):**
   - The probability of an event E, denoted as P(E), is the ratio of the number of favorable outcomes to the total number of possible outcomes in the sample space. It is calculated as:
   
   $$P(E) = \frac{\text{Number of Favorable Outcomes}}{\text{Total Number of Possible Outcomes}}$$

**7. Complementary Event (E'):**
   - The complementary event of an event E, denoted as E', consists of all outcomes in the sample space that are not part of E. In other words, it represents the event "not E."

**8. Independent Events:**
   - Two events, E1 and E2, are independent if the occurrence or non-occurrence of one does not affect the probability of the other. Mathematically, P(E1 and E2) = P(E1) * P(E2).

**9. Dependent Events:**
   - Two events, E1 and E2, are dependent if the occurrence or non-occurrence of one event affects the probability of the other. The probability of E2 occurring depends on whether E1 has occurred. 
   - *For Example*:- If you have 3 balls in a jar, 2 green and 1 blue. You are asked to pick a green ball, there will be a probability of 2/3 however if we are asked to pick the green ball again, the probability will be 1/2. This is the case of a dependent event.

**10. Mutually Exclusive Events:**
   - Mutually exclusive events cannot occur simultaneously. If one event happens, the other cannot. The probability of the union of mutually exclusive events is the sum of their individual probabilities. 
   - *For E.g* :- Getting a head or a tail on flipping the coin. We cannot get the head and the tail at the same time hence this is a mutually exclusive event

**11. Conditional Probability:**
   - Conditional probability is the probability of an event occurring given that another event has already occurred. It is denoted as P(E|F), where E is the event of interest, and F is the condition.

**12. Addition and Multiplication Rules:**
   - The addition rule calculates the probability of the union of two events, and the multiplication rule calculates the probability of the intersection of two events.
   - In probability theory, the formula for the multiplication of probabilities, often referred to as the "Multiplication Rule," which is written as follows:

**For two independent events A and B:**

$$P(A \cap B) = P(A) \cdot P(B)$$

This formula states that the probability of both event A and event B occurring (the intersection of A and B) is equal to the product of the probabilities of A and B individually.

If you have more than two independent events, you can extend the multiplication rule as follows:

**For n independent events A_1, A_2, ..., A_n:**

$$P(A_1 \cap A_2 \cap \ldots \cap A_n) = P(A_1) \cdot P(A_2) \cdot \ldots \cdot P(A_n)$$

This formula allows you to calculate the probability of the intersection of multiple independent events by multiplying their individual probabilities.

**FACTORIAL**

Factorial, denoted by the symbol "!", is a mathematical operation used to calculate the product of all positive integers from 1 up to a given positive integer n. In other words, the factorial of a non-negative integer n is the product of all the whole numbers from 1 to n.

The factorial of a non-negative integer n is typically represented as n! and is defined as:

$$n! = n × (n - 1) × (n - 2) × ... × 3 × 2 × 1$$

Here are some key points about factorials:

1. **Factorial Notation:** Factorials are denoted using the exclamation mark, e.g., 5! is read as "five factorial."

2. **Factorial of 0:** By convention, the factorial of 0 is defined as 1. In mathematical terms, 0! = 1.

3. **Factorial of 1:** The factorial of 1 is also 1, i.e., 1! = 1.

4. **Factorial of Large Numbers:** Factorials grow rapidly as the input number increases. For example, 10! = 3,628,800, and 20! is a very large number.

5. **Use Cases:** Factorials are used in various mathematical and combinatorial problems, including permutations, combinations, and calculating the number of ways to arrange objects in a sequence.

6. **Recursive Definition:** Factorials can be defined recursively as follows:
   - n! = n × (n-1)! for n > 0
   - 0! = 1 (base case)

7. **Gamma Function:** The factorial concept is extended to non-integer values through the gamma function (Γ). For non-negative integers, Γ(n) = (n-1)!, and Γ(1) = 1.

Here are a few examples of factorials:
- 3! = 3 × 2 × 1 = 6
- 4! = 4 × 3 × 2 × 1 = 24
- 5! = 5 × 4 × 3 × 2 × 1 = 120

Factorials are fundamental in various areas of mathematics, including combinatorics, calculus, and probability theory. They are used to count permutations, combinations, and solve various mathematical problems involving sequences and arrangements.

**Permutation:**

In mathematics, a permutation is an arrangement of objects in a specific order. It is a way of selecting and arranging a subset of objects from a larger set without repetition. Permutations are used to count the number of ways objects can be ordered or arranged.

There are two types of permutations:

1. **Permutations without Repetition:** In this type, each object is unique, and no object can appear more than once in a permutation. The number of permutations of n distinct objects taken r at a time is denoted as "n P r" or P(n, r). The formula for permutations without repetition is:

   $$
   P(n, r) = \frac{n!}{(n - r)!}
   $$

   For example, if you have the letters A, B, and C, there are 3P2 = 6 permutations: AB, AC, BA, BC, CA, and CB.

2. **Permutations with Repetition:** In this type, some objects may be identical, and they can be repeated in the arrangement. The number of permutations of n objects with repetition, taken r at a time, is calculated using the formula:

   $$
   P_{\text{rep}}(n, r) = n^r
   $$

   For example, if you have the letters A, A, B, and C, and you want to find 3P2 with repetition, there are 4^2 = 16 permutations.

````python
# Example 1: Find permutations of the letters A, B, and C
letters = ['A', 'B', 'C']
permutations_ABC = list(permutations(letters))
print("Permutations of ABC:")
for p in permutations_ABC:
    print("".join(p))
print(f'\nThere are {len(permutations_ABC)} permutations of the letters A, B, and C')
# Example 2: Find permutations of 4 books on a shelf
books = ['Book1', 'Book2', 'Book3', 'Book4']
permutations_books = list(permutations(books))
print("\nPermutations (without repitition) of 4 books on a shelf:\n")
for p in permutations_books:
    print(", ".join(p))
print(f'\nThere are {len(permutations_books)} permutations of 4 books on a shelf')
````

**Combination:**

A combination, on the other hand, is a selection of objects from a set without regard to the order in which they are arranged. Unlike permutations, combinations focus on choosing objects without considering their order. Combinations are used to count the number of ways to select a subset from a larger set.

There are two types of combinations:

1. **Combinations without Repetition:** In this type, each object is unique, and no object can be selected more than once in a combination. The number of combinations of n distinct objects taken r at a time is denoted as "n C r" or C(n, r). The formula for combinations without repetition is:

   $$
   C(n, r) = \frac{n!}{r!(n - r)!}
   $$

   For example, if you have the letters A, B, and C, there are 3C2 = 3 combinations: AB, AC, and BC.

   C_{\text{rep}}(n, r) = C(n + r - 1, r)
   $$

   For example, if you have the letters A, A, B, and C, and you want to find 3C2 with repetition, there are C(4, 2) = 10 combinations.

## Probability Distribution

A probability distribution is a fundamental concept in probability theory and statistics that helps us understand the likelihood of different outcomes in random events. It's like a tool we use to deal with uncertainty in a wide range of situations, from predicting coin tosses to estimating temperatures or customer arrivals.

Here are some key points:

1. **Random Variables:** Probability distributions are linked to random variables, which represent outcomes that happen by chance. These variables can stand for anything from the roll of a die to the result of a scientific experiment.

2. **Types of Distributions:** There are different types of probability distributions for different scenarios. Some common ones include the bell-shaped normal distribution, the binomial distribution (used in yes-or-no questions), and the exponential distribution (for events happening over time).

3. **Expected Value and Variance:** These are numbers we use to summarize probability distributions. The expected value is like the "average" outcome, while variance tells us how spread out the outcomes are.

4. **Applications:** We use probability distributions to make predictions and decisions in fields like finance, science, and everyday life. For example, they help with risk assessments, quality control, and understanding data.

5. **Visualizing Distributions:** We often represent probability distributions with charts like histograms or curves. These visuals make it easier to grasp what's going on.

In simple terms, probability distributions are tools that help us deal with unpredictability and make sense of uncertain situations. They're used in many areas where understanding and managing risk is important. 

Let's plot some probability distributions first.

Example 1:

````python
# Define parameters for the normal distribution
mu = 60  # Mean
sigma = 10  # Standard deviation
sample_size = 10000  # Number of data points in the sample

# Generate random data points from a normal distribution
data = np.random.normal(mu, sigma, sample_size)

# Create a histogram to visualize the distribution
plt.hist(data, bins=30, density=True, color='#219C90', alpha=0.7, label='Weight (in kgs)')

# Create a range of x values for the probability density function (PDF) of the normal distribution
x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 100)
# Calculate the PDF values for the given x values
pdf = 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-(x - mu)**2 / (2 * sigma**2))
# Plot the PDF as a curve
plt.plot(x, pdf, 'r--', label='Normal Distribution')

# Add labels and a legend
plt.xlabel('Weight (in kgs)')
plt.ylabel('Probability Density')
plt.title('Weight Probability Distribution')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.legend()

# Show the plot
plt.show()
````

**Normal Distribution:**

The normal distribution, also known as the Gaussian distribution or bell curve, is a continuous probability distribution that is symmetric and bell-shaped. It's characterized by two parameters: the mean (μ), which represents the central value, and the standard deviation (σ), which measures the spread or variability of the data.


**Properties:**

1. **Continuous Distribution:** The normal distribution is a continuous probability distribution, which means it can take any real value within a range.

2. **Symmetrical Bell Curve:** It's characterized by a symmetrical bell-shaped curve centered around the mean.

3. **Mean and Variance:** The distribution is fully defined by two parameters: the mean (μ) and the standard deviation (σ), which control the location and spread of the curve.

4. **Probability Density Function (PDF):** The PDF of the normal distribution is described by the famous Gaussian function.

5. **Central Limit Theorem:** It's often associated with the Central Limit Theorem, which states that the sampling distribution of the sample mean for a large enough sample size (n) from any population follows a normal distribution.

6. **Empirical Rule:** The empirical rule states that approximately 68% of data falls within one standard deviation of the mean, 95% within two standard deviations, and 99.7% within three standard deviations.

**Applications:** 

Many natural phenomena follow a normal distribution, including heights, IQ scores, and errors in measurements. It's widely used in statistical inference and hypothesis testing.

**Example**

````python
# Let's say we're measuring temperatures with errors, and the error follows a normal distribution.
mean_temperature = 25              # Mean temperature in degrees Celsius
std_deviation = 1.5                # Standard deviation of temperature measurement error in degrees Celsius
sample_size = 10000                # Number of temperature measurements

# Generate random data points representing measured temperatures with errors
measured_temperatures = np.random.normal(mean_temperature, std_deviation, sample_size)

# Create a histogram to visualize the distribution of measured temperatures with errors
plt.hist(measured_temperatures, bins=30, density=True, alpha=0.5, color='b', label='Measured Temperatures with Errors')

# Create a range of x values for the probability density function (PDF) of the normal distribution
x = np.linspace(mean_temperature - 3 * std_deviation, mean_temperature + 3 * std_deviation, 100)
# Calculate the PDF values for the given x values
pdf = 1 / (std_deviation * np.sqrt(2 * np.pi)) * np.exp(-(x - mean_temperature)**2 / (2 * std_deviation**2))
# Plot the PDF as a curve
plt.plot(x, pdf, 'r--', label='Normal Distribution PDF')

# Add labels and a legend
plt.xlabel('Temperature (°C)')
plt.ylabel('Probability Density')
plt.title('Measured Temperatures with Errors (Normal Distribution)')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.legend()

# Show the plot
plt.show()
````

**Binomial Distribution:**

The binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of independent Bernoulli trials (experiments with two possible outcomes: success or failure). It's characterized by two parameters: the number of trials ($n$) and the probability of success ($p$).


**Properties:**

1. **Discrete Distribution:** The binomial distribution is a discrete probability distribution, meaning it's used to model random experiments with a finite number of distinct outcomes.

2. **Two Possible Outcomes:** It models experiments with exactly two possible outcomes, often referred to as "success" and "failure."

3. **Fixed Number of Trials:** It's used when conducting a fixed number of independent trials, denoted as "$n$."

4. **Probability of Success:** The probability of success on each trial is denoted as "$p$."

5. **Probability Mass Function (PMF):** The PMF of the binomial distribution calculates the probability of getting a specific number of successes ($k$) in '$n$' trials.

6. **Mean and Variance:** The mean $(μ)$ of the binomial distribution is $μ = n*p$, and the variance $(σ^2)$ is $σ^2 = np*{(1-p)}$.

7. **Bell-Shaped for Large '$n$':** For a large number of trials ($n$) with p close to 0.5, the binomial distribution approaches a bell-shaped curve resembling the normal distribution.

**Another Exaample:**

A manufacturer has 12% defects rate in production. The buyer decides to test 20 random pieces and will accept the supplier if there are 2 or less defectives. What is the probability of getting accepted?

*SOLUTION*

To calculate the probability of the manufacturer being accepted by the buyer, given the defect rate and acceptance criteria, we can use the binomial distribution. In this case, the probability of success ($p$) is the probability of a defective item, which is 12% or 0.12, and the number of trials ($n$) is 20 (the buyer tests 20 random pieces).

We want to find the probability that there are 2 or fewer defective items ($k ≤ 2$) among the 20 tested. We can calculate this probability by summing the probabilities of having 0, 1, or 2 defective items. The probability mass function (PMF) of the binomial distribution can be used for this calculation.

The formula for the PMF of the binomial distribution is:

  $$P(X = k) = C(n,k) \cdot p^k \cdot (1 - p)^{n - k}$$

Where:

- $P(X = k)$ is the probability of getting k defective items.
- $C(n, k)$ is the binomial coefficient, which represents the number of ways to choose k items out of n.
- $p$ is the probability of success (defective item).
- $(1 - p)$ is the probability of failure (non-defective item).
- $n$ is the number of trials.

Let's calculate the probability:

1. Probability of 0 defective items (k = 0): $P(X = 0) = C(20, 0) * (0.12)^0 * (1 - 0.12)^{20-0}$

2. Probability of 1 defective item (k = 1): $P(X = 1) = C(20, 1) * (0.12)^1 * (1 - 0.12)^{20-1}$

3. Probability of 2 defective items (k = 2): $P(X = 2) = C(20, 2) * (0.12)^2 * (1 - 0.12)^{20-2}$

````python
p_defective = 0.12
n_trials = 20
# Calculate and print the probabilities
probability_0_defective = stats.binom.pmf(0, n_trials, p_defective)
probability_1_defective = stats.binom.pmf(1, n_trials, p_defective)
probability_2_defective = stats.binom.pmf(2, n_trials, p_defective)
print("Probability of 0 defective items:", round(probability_0_defective, 4))
print("Probability of 1 defective item:", round(probability_1_defective, 4))
print("Probability of 2 defective items:", round(probability_2_defective, 4))
# Calculate the probability of getting accepted (k ≤ 2)
probability_accepted = sum(stats.binom.pmf(k, n_trials, p_defective) for k in range(3))
probability_more_than_2_defectives = 1 - probability_accepted
print(f"Probability of being accepted: {round(probability_accepted, 4)} or {round(probability_accepted, 4)*100}%")
print(f'Probability of Getting more than 2 defectives: {probability_more_than_2_defectives*100:.2f}%')
````

````python
# Calculate and print the mean, standard deviation, and variance
mean = stats.binom.mean(n_trials, p_defective)
std_deviation = stats.binom.std(n_trials, p_defective)
variance = stats.binom.var(n_trials, p_defective)
print(f"Mean Of Defective Pieces: {mean:.2f}")
print(f"Standard Deviation Of Defective Pieces: {std_deviation:.2f}")
print(f"Variance Of Defective Pieces: {variance:.2f}")
````

**Poisson Distribution:**

The Poisson distribution is a discrete probability distribution that models the number of events occurring in a fixed interval of time or space when these events occur with a known constant mean rate (λ) and are independent of the time since the last event. The equation of the poisson distribution is as follows:-

$$P(X, μ) = {e^{-μ}} *\frac {μ^x}{x!}$$


**Properties**

1. **Discrete Distribution:** The Poisson distribution is a discrete probability distribution used to model rare events or the number of events that occur within a fixed interval of time or space.

2. **Single Parameter:** It's characterized by a single parameter, λ (lambda), which represents the average rate at which events occur.

3. **Probability Mass Function (PMF):** The PMF calculates the probability of observing a specific number of events (k) within the specified interval.

4. **Events Are Independent:** It assumes that events occur independently of each other within the interval.

5. **Mean and Variance:** The mean ($μ$) and variance ($σ^2$) of the Poisson distribution are both equal to λ.

6. **Skewed Distribution:** The Poisson distribution is positively skewed, with a long tail on the right side.

7. **Use Cases:** It is often used in scenarios such as modeling the number of customer arrivals at a store, the number of phone calls at a call center in a given minute, or the number of accidents at an intersection in a day.


**Example:** 

On a booking counter on the avge 3.6 people come every 10 minutes on weekends, whaat is the probability of getting 7 people in 10 minutes

````python
# Average rate of people arriving every 10 minutes
lambda_val = 3.6

# Number of people you want to calculate the probability for
k = 7

# Calculate the probability using the Poisson distribution formula
probability = (math.exp(-lambda_val) * lambda_val**k) / math.factorial(k)

# Print the result
print(f"The probability of getting  exactly {k} people in 10 minutes is approximately {probability:.4f}")
````

````python
# Average rate of people arriving every 10 minutes
lambda_val = 3.6

# Number of people you want to calculate the probability for
k = 7

for i in range(0,8):
    # Calculate the probability using the Poisson distribution formula
    probability = (math.exp(-lambda_val) * lambda_val**i) / math.factorial(i)
    # Print the result
    print(f"The probability of getting {i} people in 10 minutes : {probability:.4f}")
````

````python
colors = ['red' if i == 7 else 'lightgray' for i in range(8)]
# Average rate of people arriving every 10 minutes
lambda_val = 3.6

# Number of people you want to calculate the probability for
k = 7

# Generate a range of possible arrival counts (from 0 to 20)
arrivals = np.arange(0, 11)

# Calculate the probabilities for each arrival count using the Poisson distribution
probabilities = stats.poisson.pmf(arrivals, lambda_val)

# Create a bar plot to visualize the Poisson distribution
plt.bar(arrivals, probabilities, align='center', color=colors, alpha=0.7, label='PMF')
plt.xlabel('Number of People in 10 Minutes')
plt.ylabel('Probability')
plt.title(f'Poisson Distribution (λ={lambda_val})')
plt.xticks(arrivals)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Annotate the probability for 7 people
probability_7 = probabilities[k]
plt.annotate(f'Probability of 7 people: {probability_7:.4f}', (k, probability_7), xytext=(k + 1, probability_7 + 0.05),
             arrowprops=dict(arrowstyle='->'))

# Show the plot
plt.legend()
plt.show()
````
