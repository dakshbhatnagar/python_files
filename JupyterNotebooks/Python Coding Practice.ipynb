3. **Check if a string is a palindrome**: Write a function to determine whether a given string is a palindrome (reads the same forwards and backwards).

````python
def is_palindrome(text):
    reversed_text = text[::-1]
    return text == reversed_text
````

````python
is_palindrome('banana')
````

````python
is_palindrome('level')
````

4. **Calculate the factorial of a number**: Write a function to calculate the factorial of a given non-negative integer.

````python
def factorial(num):
    if num<=1:
        return 1
    else:
        result =  1
        # 2 because we have already defined the case for 1, num+1 because python will go to num-1 if we do just num
        for i in range(2, num+1):
            result *= i
        return result
````

````python
factorial(5)
````

5. **Reverse a string**: Write a function to reverse a given string.

````python
def reverse_string(text):
    new_text = list(str(text))
    for i in range(len(new_text)-1,0,-1):
        new_text[i], new_text[len(text) - i -1] =  new_text[len(text) - i - 1], new_text[i]
    return ''.join(new_text)
````

````python
reverse_string('daksh')
````

6. **Find the intersection of two lists**: Write a function to find the intersection of two given lists (i.e., elements common to both lists).

````python
def intersection(arr1, arr2):
    first_list = list(arr1)
    second_list =  list(arr2)
    return [i for i in first_list if i in second_list]
````

````python
intersection([1,2,3,4], [3,4,5])
````

8. **Remove duplicates from a list**: Write a function to remove duplicate elements from a given list while preserving the order of elements.

````python
def remove_duplicates(arr1):
    new_arr = []
    for i in arr1:
        if i not in new_arr:
            new_arr.append(i)
    return new_arr
````

````python
arr = [1,2,3,4,4,5,5,5]
remove_duplicates(arr)
````

9. **Sort a list of tuples based on the second element**: Write a function to sort a list of tuples based on the value of the second element in each tuple.

````python
#Creating a tuple first to sort them later
string = 'banana'
result = [(i, string.count(i)) for i in set(string)]
result
````

````python
def sort_tuples(tuples):
    #Using key argument to sort the tuples by the second element.
    return sorted(tuples, key =  lambda x : x[-1], reverse=True)
````

````python
sort_tuples(result)
````

10. **Find the missing number in a list**: Write a function to find the missing number in a given list of consecutive integers.

````python
arr =  [1,2,3,4,6,7,8]
````

````python
def find_missing(arr):
    MAX = max(arr)
    MIN =  min(arr)
    missing = [i for i in range(MIN,MAX) if i not in arr]
    return missing
````

````python
find_missing(arr)
````

Can you use groupby and apply to perform custom aggregation on a DataFrame?

````python
def custom_agg(series):
    mean, count = round(series.mean(),2), series.count().apply(lambda x : int(x))
    return pd.Series([mean[0], count[0]], index=['Mean_Sales', 'Count_Observations'])
````

````python
pd.DataFrame(df.groupby(['Segment', 'Region'])[['Sales']].apply(custom_agg))
````
